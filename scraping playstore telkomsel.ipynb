```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
#Scrape desired number of reviews

from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.telkomsel.telkomselcm',
    lang='id', # defaults to 'en'
    country='id', 
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT you can use Sort.NEWEST to get newst reviews
    count=250000, # defaults to 100
    filter_score_with=None # defaults to None(means all score) Use 1 or 2 or 3 or 4 or 5 to select certain score
)
```

```python
df_telkomsel = pd.DataFrame(np.array(result),columns=['review'])
df_telkomsel = df_telkomsel.join(pd.DataFrame(df_telkomsel.pop('review').tolist()))
df_telkomsel.head(250000)
```

```python
len(df_telkomsel.index) #count the number of data we got
```

```python
#Run This Code to Sort the Data By Date 

new_df = df_telkomsel[['userName', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False) #Sort by Newst, change to True if you want to sort by Oldest.
sorted_df.head(250000)
```

```python
sorted_df.to_excel('data_playstore_mytelkomsel.xlsx')
```
