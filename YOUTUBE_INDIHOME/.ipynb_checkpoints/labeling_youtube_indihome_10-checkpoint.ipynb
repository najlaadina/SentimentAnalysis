```python
import pandas as pd
from textblob import TextBlob
```

```python
columns = ["comment_english"]
df = pd.read_excel(r'C:\Users\Hp\TUGAS AKHIR 2\YOUTUBE_INDIHOME\databersih_english_youtube_indihome\databersih_english_youtube_indihome_10.xlsx', usecols=columns)
df.head(2000)
```

```python
df['comment_english'] = df['comment_english'].str.lower()
df.head(2000)
```

```python
df['label_awal'] = ''
for i,x in df.comment_english.iteritems():
    label_awal = TextBlob(x)
    df['label_awal'][i] = label_awal.sentiment.polarity
    print("Index: ", i ,  "label_awal" , label_awal.sentiment.polarity)
```

```python
df.head()
```

```python
df['label_awal'].value_counts()
```

```python
label = []
for index, row in df.iterrows():
    if row['label_awal'] < 0.0:
        label.append("negative")
    else:
        label.append("positive")
df["label"] = label
df = df.drop(columns=['label_awal'])
df.head(2000)
```

```python
df['label'].value_counts()
```

```python
df.to_excel('labeling_databersih_youtube_indihome_10.xlsx')
```
